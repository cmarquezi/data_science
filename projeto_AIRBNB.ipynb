## Q4. Verificando se há Outliers presente no dataframe

Uma forma que auxilia na identificação de Outliers e usar as funções DESCRIBE e BOXPLOT. Desta forma pode verificar numericamente ou por gráfico. Numericamente, tem que
observar a Mediana e o Valor Máximo. Se estiver distorcido, é um indicativo de Outliers. Já através de gráfico, como o boxplot, os pontos que esá muito distante dos demais
também é um indicio de outliers.

Vamos como usando a função DESCRIBE, para as columas núméricas.


```python
# Usando o Método Describe para alguma variável especifica

'''
Formas para usar a função Describe para uma variável
 df_rio.price.describe()
 df_rio['price'].describe()'''

# Describe de alguns parâmetros antes para avaliar se existem Outliers
df_rio[['price', 'minimum_nights', 'number_of_reviews','reviews_per_month', 'calculated_host_listings_count', 'availability_365']].describe()

# Podemos observar como na coluna "price", o valor máximo está muito acima da média, logo, é um indicativo de Outliers
# A mesma coisa para "minimum_night".
# A mesma reflexão vale as demais variáveis.
```

Como observamos acima, a coluna PRICE, tem média 731,18. E se observar, 75% dos valores estão abaixo de de 572,00. Logo, o valor máximo de valor máximo de 633.125,00, indica Outliers

Vamos então usar o boxplot para comprovar por meio de gráfico

A mesma analise podemos fazer para a coluna MINIMUM_NIGHT. A média é de 3,68 noites, e 75% dos valores abaixo de 3 noites. Já o valor máximo 1000.

```python
# Plotando histograma antes da limpeza
df_rio.hist(['price', 'minimum_nights'], bins=15, figsize=(10,3))
plt.show()
```

```python
# Abaixo o boxplot da variável "price"

df_rio.boxplot(column=['price'], vert=False, grid=False,figsize=(15,3))
plt.show()

# Verificando a quantidade de valores acima os 75% para 'price'

print(' price: valores acima de R$ 1.500,00:')
print(f' Quantas Entradas acima de R$ 1.500,00: {len(df_rio[df_rio.price > 1500])} de {df_rio.shape[0]}')
print(f' Percentual: {len(df_rio[df_rio.price > 1500]) / df_rio.shape[0] * 100 :.2f}%')
```

## Q.4 Qual a correlação existente entre as variáveis

A correlação entre as variáveis é uma medida estatística que descreve a relação entre duas coisas.

Essa relação pode ser Positiva, Negativa ou Neutra.

Para isso podemos o usar o modulo do Pandas para encontra a correlação das variáveis númericas:

[click me ] >> https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html#



```python
# Novamente pode localizar a variáveis através do metodo Columns, aplicado do DF_CLEAN
df_clean.columns
```

```python
corr = df_clean[['price', 'minimum_nights',
       'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365']].corr()
```

```python
# Se usar diretamento CORR, como abaixo, aparece apenas uma tabela. Pode usar de maneira gráfica para facilitar o entendimento

display(corr)
```

```python
# Como os dados acima gerados, vamos criar um grafico para demonstra a correlação entre as variáveis

''' SEABORN
https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn.heatmap '''

sns.heatmap(corr, annot=True, fmt='.2g', linecolor='white', square=True, cmap='BrBG')
plt.show()
```

```python
# identificando os outliers para a variavel Price

q1_price = df_rio.price.quantile(.25)
q3_price = df_rio.price.quantile(.75)
IQR_price = q3_price - q1_price
print('IQR da variável price: ', IQR_price)

# definindo os limites
sup_price = q3_price + 1.5 * IQR_price
inf_price = q1_price - 1.5 * IQR_price

print('Limite Superior de Price: ', sup_price)
print('Limite Inferior de Price: ', inf_price)
```

```python
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df_rio.price.plot(kind='box', vert=False);
ax.set_title('Dataset Original - price')
plt.show()
print("O dataset possui {} colunas".format(df_rio.shape[0]))
print("{} Entradas acima de 1131,5".format(len(df_rio[df_rio.price > 1131.5])))
print("Representam {:.2f}% do dataset".format((len(df_rio[df_rio.price > 1131.5]) / df_rio.shape[0])*100))
```

```python
# identificando os outliers para a variavel minimum_nights

q1_minimum_nights = df_rio.minimum_nights.quantile(.25)
q3_minimum_nights = df_rio.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('IQR da variável price: ', IQR_minimum_nights)

# definindo os limites
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights

print('Limite Superior de minimum_nights: ', sup_minimum_nights)
print('Limite Inferior de minimum_nights: ', inf_minimum_nights)
```

```python
fig, ax = plt.subplots(figsize=(15,3))
df_rio.minimum_nights.plot(kind='box', vert=False);
ax.set_title('Dataset Original - minimum_nights')
plt.show()
print("O dataset possui {} colunas".format(df_rio.shape[0]))
print("{} Entradas acima de 6".format(len(df_rio[df_rio.minimum_nights > 6])))
print("Representam {:.2f}% do dataset".format((len(df_rio[df_rio.minimum_nights > 6]) / df_rio.shape[0])*100))
```

```python
# Limpando o dataset

df_clean = df_rio.copy()

df_clean.drop(df_clean[df_clean.price > 1131.5].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 6].index, axis=0, inplace=True)

# Tamanhao do dataset antes e depois

print('Antes da limpeza: ',  df_rio.shape[0])
print('Depois da limpeza: ', df_clean.shape[0])
```
